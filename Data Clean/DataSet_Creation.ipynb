```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

random.seed(9)
np.random.seed(9)

n_rows = 10000

first_names = ['Mike', 'Teddy', 'Uma', 'Jane', 'Uttam', 'Kim', 'Raj', 'Anita', 'Sam',
               'Priya', 'Carlos', 'Linda', 'Tom', 'Nina', 'Omar', 'Zara', 'Leo', 'Maya',
               'Ravi', 'Sara', 'John', 'Asha', 'Dev', 'Fatima', 'George', 'Aman', 'Pooja',
               'Vikram', 'Neha', 'Rohan', 'Divya', 'Arjun', 'Kavya', 'Sahil', 'Meera',
               'Karan', 'Riya', 'Aditya', 'Simran', 'Yusuf', 'Tara', 'Nikhil', 'Isha',
               'Aryan', 'Pallavi', 'Varun', 'Shreya', 'Ajay', 'Komal', 'Imran', 'Sneha',
               'Ethan', 'Olivia', 'Liam', 'Sophia', 'Noah', 'Emma', 'Mason', 'Ava',
               'Lucas', 'Mia', 'Daniel', 'Grace', 'Henry', 'Chloe', 'Aarav', 'Diya',
               'Ishaan', 'Anaya', 'Rehan', 'Kiara', 'Vivaan', 'Myra', 'Reyansh', 'Aadhya',
               'Krishna', 'Saanvi', 'Aryaman', 'Avni', 'Dhruv', 'Anvi', 'Kabir', 'Ira',
               'Veer', 'Navya', 'Shaurya', 'Disha', 'Ranbir', 'Trisha', 'Aniket', 'Manya',
               'Harsh', 'Vidya', 'Akash', 'Tanvi', 'Ishan', 'Rhea', 'Mihir', 'Pari',
               'Oliver', 'Amelia', 'Elijah', 'Harper', 'James', 'Evelyn', 'Benjamin', 'Abigail',
               'William', 'Ella', 'Alexander', 'Scarlett', 'Michael', 'Lily', 'Sebastian', 'Hannah']

last_names = ['Sharma', 'Verma', 'Iyer', 'Patel', 'Khan', 'Reddy', 'Nair', 'Gupta',
              'Mehta', 'Joshi', 'Singh', 'Kapoor', 'Chopra', 'Das', 'Bose', 'Pillai',
              'Rao', 'Desai', 'Malhotra', 'Bhatt', 'Agarwal', 'Saxena', 'Mukherjee',
              'Sinha', 'Trivedi', 'Bansal', 'Chauhan', 'Dube', 'Ghosh', 'Kulkarni',
              'Menon', 'Naidu', 'Pandey', 'Rana', 'Shah', 'Thakur', 'Yadav', 'Anand',
              'Bhalla', 'Chandra', 'Dutta', 'Garg', 'Hegde', 'Jain', 'Kohli', 'Lal',
              'Mathur', 'Nambiar', 'Ojha', 'Prasad', 'Rastogi', 'Sengupta', 'Taneja',
              'Upadhyay', 'Varma', 'Walia', 'Zachariah', 'Smith', 'Johnson', 'Brown',
              'Williams', 'Miller', 'Davis', 'Wilson', 'Clark', 'Lewis', 'Walker',
              'Bhatia', 'Chakraborty', 'Dixit', 'Easwaran', 'Farooqui', 'Goel', 'Hari',
              'Iqbal', 'Jha', 'Kaur', 'Lamba', 'Madan', 'Narayan', 'Oberoi', 'Puri',
              'Qureshi', 'Raina', 'Solanki', 'Tandon', 'Uppal', 'Vaidya', 'Wadhwa',
              'Xavier', 'Yusufzai', 'Zaidi', 'Acharya', 'Bedi', 'Chhabra', 'Deshpande',
              'Eswaran', 'Fernandes', 'Gokhale', 'Hooda', 'Israni', 'Jindal']

domains = ['Datascience', 'Testing', 'Dataanalyst', 'Analytics', 'Statistics', 'NLP',
           'MachineLearning', 'WebDevelopment', 'DevOps', 'CloudEngineering']

cities = ['Mumbai', 'Bangalore', 'Hyderabad', 'Delhi', 'Pune', 'Chennai', 'Kolkata', 'Ahmedabad']
city_typos = {'Hyderabad': 'Hyderbad', 'Bangalore': 'Banglore', 'Ahmedabad': 'Ahmadabad'}

departments = ['Engineering', 'Sales', 'HR', 'Finance', 'Operations', 'Marketing', 'Support']
genders = ['Male', 'Female']
marital_statuses = ['Single', 'Married', 'Divorced']
education_levels = ['Bachelors', 'Masters', 'PhD', 'Diploma']
employee_statuses = ['Active', 'Resigned', 'Terminated']

SYMBOLS = ['^', '#', '*', '$', '!', '&', '%', '+']


def unique_full_name(used_names, max_attempts=8):
    """Build a first+last combo, retrying a bounded number of times to avoid
    collisions. With a few thousand possible combos and 10,000 rows, a small
    number of repeats is expected and realistic (real datasets have name
    repeats too) -- this just minimizes them without risking an infinite loop."""
    for _ in range(max_attempts):
        full = f"{random.choice(first_names)} {random.choice(last_names)}"
        if full not in used_names:
            used_names.add(full)
            return full
    # pool exhausted for this attempt budget -- accept a repeat
    full = f"{random.choice(first_names)} {random.choice(last_names)}"
    used_names.add(full)
    return full


def messy_name(name):
    if random.random() < 0.30:
        sym = random.choice(SYMBOLS)
        if random.random() < 0.5:
            name = name + sym
        else:
            mid = len(name) // 2
            name = name[:mid] + sym + name[mid:]
    if random.random() < 0.05:
        name = ' ' + name + '  '   # stray whitespace
    return name


def messy_domain(d):
    out = d
    if random.random() < 0.45:
        out += random.choice(['#$', '^^', '^^#', '!!', '**', '##'])
    if random.random() < 0.25:
        out = out.lower()
    elif random.random() < 0.1:
        out = out.upper()
    return out


def messy_age():
    if random.random() < 0.15:
        return np.nan
    age = random.randint(22, 65)
    suffix = random.choice([' years', "' yr", '-yr', 'yr', ' yr', ''])
    return f"{age}{suffix}"


def messy_location():
    if random.random() < 0.15:
        return np.nan
    city = random.choice(cities)
    if random.random() < 0.12 and city in city_typos:
        city = city_typos[city]
    return city


def messy_salary():
    if random.random() < 0.06:
        return np.nan
    base = random.choice([5, 6, 8, 10, 12, 15, 18, 20, 25, 30, 35, 40, 50, 60, 75, 90, 100]) * 1000
    s = str(base)
    if random.random() < 0.55:
        pos = random.randint(1, len(s) - 1)
        sym = random.choice(['^', '%', '$', '#'])
        s = s[:pos] + sym + s[pos:]
    return s


def messy_exp():
    if random.random() < 0.15:
        return np.nan
    years = random.randint(1, 15)
    fmt = random.choice([
        f"{years}+",
        f"<{years}",
        f"{years}> yrs",
        f"{years}+ year",
        f"{years} yrs",
        str(years),
    ])
    return fmt


def messy_department(dept):
    out = dept
    if random.random() < 0.2:
        out = out.lower()
    elif random.random() < 0.1:
        out = out.upper()
    if random.random() < 0.15:
        out += random.choice(['_dept', ' Dept.', '##'])
    if random.random() < 0.08:
        return np.nan
    return out


def random_messy_date(start_year, end_year):
    start = datetime(start_year, 1, 1)
    end = datetime(end_year, 12, 31)
    d = start + timedelta(days=random.randint(0, (end - start).days))
    fmt = random.choice([
        '%Y-%m-%d',     # 2021-05-04
        '%d/%m/%Y',     # 04/05/2021
        '%d-%b-%Y',     # 04-May-2021
        '%b %d, %Y',    # May 04, 2021
        '%d.%m.%Y',     # 04.05.2021
        '%Y/%m/%d',     # 2021/05/04
    ])
    text = d.strftime(fmt)
    if random.random() < 0.05:
        text = text + ' '   # stray whitespace
    return d, text


def messy_gender(g):
    variants = {
        'Male': ['Male', 'M', 'm', 'MALE', 'male '],
        'Female': ['Female', 'F', 'f', 'FEMALE', 'female '],
    }
    if random.random() < 0.1:
        return np.nan
    return random.choice(variants[g])


def messy_marital_status(ms):
    out = ms
    if random.random() < 0.3:
        out = out.upper() if random.random() < 0.5 else out.lower()
    if random.random() < 0.1:
        out = out[0]   # abbreviate e.g. 'S', 'M', 'D'
    if random.random() < 0.12:
        return np.nan
    return out


def messy_performance_rating():
    if random.random() < 0.18:
        return np.nan
    rating = round(random.uniform(1.0, 5.0), 1)
    fmt = random.choice([
        str(rating),
        f"{rating}/5",
        f"{int(rating * 20)}%",
        f"{rating} stars",
        f"{rating}^",
    ])
    return fmt


def messy_email(name):
    base = ''.join(ch for ch in name if ch.isalnum()).lower()
    if not base:
        base = 'user'
    domain_part = random.choice(['gmail.com', 'company.com', 'outlook.com', 'yahoo.in'])
    email = f"{base}{random.randint(1,99)}@{domain_part}"
    if random.random() < 0.08:
        email = email.replace('@', '_at_')          # malformed
    if random.random() < 0.08:
        email = email.replace('.com', '.con')        # typo
    if random.random() < 0.06:
        email = ' ' + email                           # stray whitespace
    if random.random() < 0.1:
        return np.nan
    return email


def messy_phone():
    if random.random() < 0.12:
        return np.nan
    num = ''.join(str(random.randint(0, 9)) for _ in range(10))
    fmt = random.choice([
        num,
        f"+91-{num}",
        f"{num[:5]} {num[5:]}",
        f"(+91) {num}",
        f"0{num}",
        f"{num[:3]}-{num[3:6]}-{num[6:]}",
    ])
    return fmt


def messy_bonus():
    if random.random() < 0.2:
        return np.nan
    base = random.choice([0, 1000, 2000, 2500, 5000, 7500, 10000, 15000, 20000])
    s = str(base)
    if random.random() < 0.4:
        pos = random.randint(0, len(s) - 1)
        sym = random.choice(['^', '%', '$', '#'])
        s = s[:pos] + sym + s[pos:]
    return s


def messy_attendance():
    if random.random() < 0.15:
        return np.nan
    pct = round(random.uniform(60, 100), 1)
    fmt = random.choice([
        f"{pct}%",
        f"{pct}",
        f"{int(pct)} %",
        f"{pct}pct",
    ])
    return fmt


def messy_project_count():
    if random.random() < 0.1:
        return np.nan
    n = random.randint(0, 12)
    fmt = random.choice([str(n), f"{n} projects", f"{n}+", f" {n}"])
    return fmt


def messy_remote_work():
    if random.random() < 0.1:
        return np.nan
    return random.choice(['Yes', 'No', 'Y', 'N', 'yes', 'no', 'YES ', 'NO'])


def messy_manager(used_names_list):
    if random.random() < 0.12:
        return np.nan
    mgr = random.choice(used_names_list)
    if random.random() < 0.2:
        mgr = mgr.lower()
    if random.random() < 0.1:
        mgr += random.choice(SYMBOLS)
    return mgr


def messy_education(level):
    out = level
    if random.random() < 0.25:
        out = {'Bachelors': "B.Tech/B.E", 'Masters': "M.Tech/MBA",
                'PhD': "Ph.D.", 'Diploma': "Diploma"}[level]
    if random.random() < 0.15:
        out = out.lower()
    if random.random() < 0.1:
        return np.nan
    return out


def messy_status(status):
    out = status
    if random.random() < 0.3:
        out = out.upper() if random.random() < 0.5 else out.lower()
    if random.random() < 0.08:
        return np.nan
    return out


used_names = set()
clean_names = [unique_full_name(used_names) for _ in range(n_rows)]
# a small pool of "manager-eligible" clean names, drawn from the same generated set
manager_pool = random.sample(clean_names, k=min(300, len(clean_names)))

rows = []
for i in range(n_rows):
    name = clean_names[i]
    g = random.choice(genders)
    join_dt, join_text = random_messy_date(2014, 2024)
    review_start_year = max(join_dt.year, 2023)
    _, review_text = random_messy_date(review_start_year, 2025)
    rows.append([
        f"EMP{1000+i}",
        messy_name(name),
        messy_domain(random.choice(domains)),
        messy_department(random.choice(departments)),
        messy_age(),
        messy_gender(g),
        messy_marital_status(random.choice(marital_statuses)),
        messy_location(),
        join_text,
        messy_salary(),
        messy_exp(),
        messy_performance_rating(),
        messy_email(name),
        messy_phone(),
        messy_bonus(),
        messy_attendance(),
        messy_project_count(),
        messy_remote_work(),
        messy_manager(manager_pool),
        messy_education(random.choice(education_levels)),
        review_text,
        messy_status(random.choice(employee_statuses)),
    ])

columns = ['Employee ID', 'Name', 'Domain', 'Department', 'Age', 'Gender',
           'Marital Status', 'Location', 'Join Date', 'Salary', 'Exp',
           'Performance Rating', 'Email', 'Phone Number', 'Bonus',
           'Attendance_Percentage', 'Project Count', 'Remote Work',
           'Manager Name', 'Education Level', 'Last Review Date',
           'Employee Status']

messy_df = pd.DataFrame(rows, columns=columns)

# inject some exact duplicate rows (common real-world issue)
dup_idx = messy_df.sample(n=6, random_state=1).index
messy_df = pd.concat([messy_df, messy_df.loc[dup_idx]], ignore_index=True)

# inject a couple of fully-missing rows
for _ in range(2):
    messy_df.loc[len(messy_df)] = [np.nan] * len(columns)

# shuffle rows so duplicates/blank rows aren't all at the bottom
messy_df = messy_df.sample(frac=1, random_state=2).reset_index(drop=True)

messy_df.to_csv('raw_employee_data.csv', index=False)
print(f"Messy dataset generated! Shape: {messy_df.shape}")
print(f"Unique names (before messiness applied): {len(set(clean_names))}")
print(messy_df.head(8).to_string())
```
